```python
import tabula
import pandas as pd
```

```python
pdf_01  =  "Comparativo por Ano.pdf"
dfs  =  tabula.read_pdf ( 
    pdf_01 , 
    pages = "all" , 
    lattice = True , 
    pandas_options = { "header" :  [ 0 ]}, 
    area = [ 0 ,  0 ,  100 ,  100 ], 
    relative_area = True , 
    multiple_tables = False , 
) 
dfs [ 0 ]
```

```python
df = pd.DataFrame(dfs[0])
```

```python
new_header = df.iloc[0] # pega a primeira linha para o cabeçalho
df = df[1:] #captura os dados menos a linha do cabecalho
df.columns = new_header #definir a linha do cabeçalho como o cabeçalho df
```

```python
df
```

```python
df.drop(27, axis = 0, inplace = True)
df.drop(32, axis = 0, inplace = True)
```

```python
df.rename(columns={'Total\rGeral': 'Total Geral'}, inplace = True)
df['Ano'].replace('Total\rGeral', 'Total Geral', inplace = True)
```

```python
df = df.rename({'Qtde\rApres.': 'Quantidade Apresentada'}, axis = 1)
df = df.rename({'Qtde\rAprov.': 'Quantidade Aprovada'}, axis = 1)
df = df.rename({'Qtde\rApoio.': 'Quantidade Apoiada'}, axis = 1)
```

```python
df.to_csv('Comparativodosprojetos.csv', sep=';', index = [0], encoding='utf-8')
```
